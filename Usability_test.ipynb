```python
import numpy as np
import matplotlib.pyplot as plt
```

# reversability test

```python
num_points = 100000

x = np.random.normal(0,5,num_points)
x = np.abs(x)

y = np.random.normal(0,2,num_points)
y = -1 * np.abs(y)

z = np.concatenate((x,y))

stdv = np.std(z)
mean = np.average(z)
print(f"Distribution qualities: mu = {mean}, sigma = {stdv}")

# cut off each side
pos = z[z >= 0]
mirror_pos = np.concatenate((pos, -1 * pos))

pos_stdv = np.std(mirror_pos)
pos_mean = np.average(mirror_pos)
print(f"positive distribution qualities: mu = {pos_mean}, sigma = {pos_stdv}")

# cut off each side
neg = z[z <= 0]
mirror_neg = np.concatenate((neg, -1 * neg))

neg_stdv = np.std(mirror_neg)
neg_mean = np.average(mirror_neg)
print(f"negative distribution qualities: mu = {neg_mean}, sigma = {neg_stdv}")
```
